```python
import argparse
import os
import time
import cv2
import torch
import numpy as np

from detectron2.config import get_cfg
from detectron2.data.detection_utils import read_image
from detectron2.utils.logger import setup_logger
from demo.predictor import VisualizationDemo
```

```python
# constants
WINDOW_NAME = "secondhands detections"
DETECTRON_DIR = "/home/jingwen/Vision/detectron2"
TEST_DIR = "/media/jingwen/Data/secondhands/train/complete_dataset/test"
IMAGE_DIR = "/media/jingwen/Data/secondhands/processed/dynamic-mallet-plier-brush-spraybottle-cutter-lab-desk-1/frames"

def setup_cfg():
    # load config from file and command-line arguments
    cfg = get_cfg()
    cfg.merge_from_file(os.path.join(DETECTRON_DIR, "configs/COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
    cfg.MODEL.ROI_HEADS.NUM_CLASSES = 5
    # cfg.DATASETS.TEST = ("secondhands_test")
    # Set score_threshold for builtin models
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
    cfg.MODEL.PANOPTIC_FPN.COMBINE.INSTANCES_CONFIDENCE_THRESH = 0.4
    cfg.OUTPUT_DIR = "/media/jingwen/Data/secondhands/train/complete_dataset/results"
    cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
    cfg.freeze()
    return cfg

def getBoundingBoxes(predictions):
    # return bounding boxes sorted according to classes order
    instances = predictions['instances']
    boxes = instances.pred_boxes
    classes = instances.pred_classes
    boxes = boxes.tensor.to(torch.device("cpu")).numpy()
    classes = classes.to(torch.device("cpu")).numpy()

    Boxes = np.zeros([5, 4])
    for i in range(0, boxes.shape[0]):
        Boxes[classes[i], :] = boxes[i, :]
    return Boxes

logger = setup_logger()
cfg = setup_cfg()
demo = VisualizationDemo(cfg)
```

```python
assoc_file = os.path.join(IMAGE_DIR, "associations.txt")
out_dir = os.path.join(IMAGE_DIR, "masks")
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
    
f = open(assoc_file)
N = len([im for im in os.listdir(IMAGE_DIR) if im.startswith("Color")])
Boxes = np.zeros((N, 5, 4))
i = 0
for line in f:
    names = line.strip("\n").split(" ")
    print(names)
    img = names[1]
    img_path = os.path.join(IMAGE_DIR, img)
    # use PIL, to be consistent with evaluation
    image = read_image(img_path, format="BGR")
    start_time = time.time()
    predictions, visualized_output = demo.run_on_image(image)
    out_filename = os.path.join(out_dir, img.replace("Color", ""))
    visualized_output.save(out_filename)
    bb = getBoundingBoxes(predictions)
    Boxes[i, :, :] = bb
    i += 1
    logger.info(
        "{}: detected {} instances in {:.2f}s".format(
            img_path, len(predictions["instances"]), time.time() - start_time
        )
    )
```

```python
np.save("bb", Boxes)
```
